```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Simple look at the dataset
df = pd.read_csv('CC GENERAL.csv')
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mean(),inplace=True)
```

```python
df_heatmap = df.drop(['CUST_ID'],axis=1)
fig, ax = plt.subplots(figsize=(10,10))
corr = df_heatmap.corr()
sns.heatmap(corr)
plt.show()
```

```python
#Lets find out which column have the highest corr with MINIMUM_PAYMENTS
corr['MINIMUM_PAYMENTS'].abs().sort_values(ascending=False)
#the BALANCE column have the highest corr with the column MINIMUM_PAYMENTS
```

```python
#Now let look at the BALANCE and MINIMUM_PAYMENTS scatter plot
fig, ax = plt.subplots(figsize=(10,10))
plt.scatter(df['BALANCE'],df['MINIMUM_PAYMENTS'])
plt.xlabel('BALANCE')
plt.ylabel('MINIMUM_PAYMENTS')
plt.title('Minimum payments based on balance')
plt.show()
```

```python
#Filter out dataset
train = df[df['MINIMUM_PAYMENTS'].notnull()]
test = df[df['MINIMUM_PAYMENTS'].isnull()]

x_train = train[['BALANCE']]
y_train = train['MINIMUM_PAYMENTS']

x_test = test[['CUST_ID','BALANCE']]
```

```python
#using simple linear regression to predict the missing value
from sklearn.linear_model import LinearRegression
reg_model = LinearRegression()
reg_model.fit(x_train,y_train)
```

```python
x_test['MINIMUM_PAYMENTS'] = reg_model.predict(x_test.drop(['CUST_ID'],axis=1))
```

```python
x_test.drop(['BALANCE'],axis=1,inplace=True)
```

```python
#Now the Nan value are filled in
df1 = df.fillna(x_test)
df1.isnull().sum()
```

```python
#Drop useless column
df1.drop(['CUST_ID'],axis=1,inplace=True)
```

```python
#Not gonna remove outlier, cuz want to capture the true scenario
df1.describe()
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
stdscaler = StandardScaler()
df_scaled = stdscaler.fit_transform(df1)
df_scaled = pd.DataFrame(df_scaled)
df_scaled.head()
```

```python
#Dimensionality reduction
from sklearn.decomposition import PCA
pca = PCA().fit(df_scaled)

plt.rcParams["figure.figsize"] = (12,6)

fig, ax = plt.subplots()
xi = np.arange(1, 18, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 18, step=1)) #change from 0-based array index to 1-based human-readable label
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% cut-off threshold', color = 'red', fontsize=16)

ax.grid(axis='x')
plt.show()
```

```python
df_dr = PCA(n_components=11).fit_transform(df_scaled)
```

```python
from scipy.cluster import hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(df_dr, method='ward'))
plt.title('Dendrogram')
plt.ylabel('Distance')
plt.show()
#From the picture we can see the best n_clusters is 2
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(df_dr)
cluster = y_hc.astype('float64')
#There have just two class, which is 1 and 0
```

```python
df_dr = pd.DataFrame(df_dr)
df_dr['CLASS'] = cluster
df_dr.head()
```

```python
#Now try to visualize the class and see how they are clustered
class_0 = df_dr[df_dr['CLASS']==0]
class_1 = df_dr[df_dr['CLASS']==1]

plt.scatter(class_0[0],class_0[1],c='b',label='Class 0')
plt.scatter(class_1[0],class_1[1],c='r',label='Class 1')
plt.title('Cluster based on feature 1 and 0')
plt.legend()

plt.tight_layout
plt.show()
```
